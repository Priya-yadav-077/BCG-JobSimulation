```python
for col in df2.select_dtypes(include=[np.number]):
  plt.hist(df2[col])
  plt.title(f"distribution:{col}")
  plt.xlabel(col)
  plt.ylabel("frequency")
  plt.show()
```

```python
for col in df1.select_dtypes(include=[np.number]):
  plt.hist(df1[col])
  plt.title(f"distribution:{col}")
  plt.xlabel(col)
  plt.ylabel("frequency")
  plt.show()
```

```python
for col in df1.select_dtypes(include=[np.number]):
  plt.plot(df1[col])
  plt.title(f"distribution:{col}")
  plt.xlabel(col)
  plt.ylabel("frequency")
  plt.show()
```

```python
for col in df1.select_dtypes(include=[object]):
  df1[col].value_counts().plot(kind='bar')
  plt.title(f"distribution:{col}")
  plt.xlabel(col)
  plt.ylabel("frequency")
  plt.show()
```

```python
for col in df2.select_dtypes(include=[object]):
  df2[col].value_counts().plot(kind='bar')
  plt.title(f"distribution:{col}")
  plt.xlabel(col)
  plt.ylabel("frequency")
  plt.show()
```
